```python
closing_price = df['Close']
```

```python
closing_price.isnull().sum()
```

```python
closing_price.dropna(inplace=True)
```

```python
closing_price.isnull().sum()
```

**For time series data, datapoint (closing price) is dependednt on existing data points (existing closing price for few days).**

So while doing train test split, we need to select data in sequence; rather than a random picking

```python
print(closing_price.shape)

print(len(closing_price))
```

```python
int(closing_price.shape[0]*0.65)
```

```python
train_size = int(closing_price.shape[0]*0.65)
test_size = len(closing_price)-train_size
```

```python
print("Training size", train_size)
print("Testing size", test_size)
```

```python
train_data = closing_price[:train_size]
test_data = closing_price[train_size:len(closing_price)]
```

```python
test_data.tail(3)
```

**Now we have to convert the prices in a way that first 5 closing prices (Monday-Friday) will be used to predict the closing proce of next week Monday**

e.g. as below

```python
closing_price[:6]
```

Here 0 to 4 index are independent features, used to predict 5th index feature

**we will write some data-processing logic to convert the closing_price series in a dataframe which is having 5 independant features and one dependent (o/p) feature**

```python
start_indx = 0
end_indx = 6

train_df = []

while 1==1:
    
    if end_indx >= train_size:
        break
    
    train_df.append(train_data[start_indx:end_indx].values)
        
    start_indx=start_indx+1
    end_indx=end_indx+1
            
train_df = [list(i) for i in train_df]
```

```python
train_df
```

```python
train_df[-1]
```

```python
train_data.tail(7)
```

```python
test_data
```

```python
start_indx = 0
end_indx = 6

test_df = []

while 1==1:
    
    if end_indx >= test_size:
        break
    
    test_df.append(test_data[start_indx:end_indx].values)
        
    start_indx=start_indx+1
    end_indx=end_indx+1
            
test_df = [list(i) for i in test_df]
```

```python
test_df
```

```python
test_df[-1]
```

```python
test_data.tail(7)
```

```python
df_train = pd.DataFrame(train_df,columns=['Close1','Close2','Close3','Close4','Close5','ClosingPrice'])
```

```python
df_train.head()
```

```python
df_train.shape
```

```python
df_test=pd.DataFrame(test_df,columns=['Close1','Close2','Close3','Close4','Close5','ClosingPrice'])
```

```python
df_test.shape
```

**we will do scaling now. As we have training and testing data available as per requirement**

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

scaled_train_data = scaler.fit_transform(df_train)

scaled_test_data = scaler.fit_transform(df_test)
```
